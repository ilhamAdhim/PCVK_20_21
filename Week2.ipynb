```
black_img = np.zeros(shape=(512,512,3), dtype=np.int16)
plt.imshow(black_img)
```

```
# Green rectangle
cv.rectangle(black_img, pt1=(384,0), pt2=(510,150), color=(0,255,0), thickness = 10)
# Center Blue rectangle
cv.rectangle(black_img, pt1=(200,200), pt2=(300,300), color=(0,0,255), thickness = 15)
# Left top circle
cv.circle(black_img, center=(100,100), radius=50, color=(255,0,0), thickness = 8)
plt.imshow(black_img)
```

```
# Make 2nd circle
cv.circle(black_img, center=(400,400), radius=50, color=(255,0,0), thickness = -1)
# Draw the line
cv.line(black_img, pt1=(0,0), pt2=(512,512), color=(255,255,255), thickness = 5)
plt.imshow(black_img)
```

```
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(black_img, text='Helo', org=(10,500), fontFace=font, fontScale=4, color=(255,255,0), thickness=2, lineType=cv.LINE_8)
plt.imshow(black_img)
```

```
class MyImage:
    def __init__(self, img_name):
        self.img = cv.imread(img_name)
        self.__name = img_name.split('/')[5]

    def name(self):
        return self.__name

src = MyImage('/content/drive/MyDrive/PCVK/kucing_lucu.jpg')
# extract blue channel
blue_channel = src.img[:,:,0]

# create empty image with same shape as that of src image
blue_img = np.zeros(src.img.shape)

#assign the red channel of src to empty image
blue_img[:,:,0] = blue_channel

cv.rectangle(src.img, pt1=(10,350), pt2=(100,250), color=(0,0,250), thickness = 5)

font = cv.FONT_ITALIC
cv.putText(src.img, text=src.name(), org=(400,300), fontFace=font, fontScale=1, color=(255,0,0), thickness=6)
plt.imshow(src.img)
```
